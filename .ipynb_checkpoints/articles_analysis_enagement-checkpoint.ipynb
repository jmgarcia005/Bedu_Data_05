```python
#Import packages from our vitualenv

import numpy as no
import pandas as pd
import matplotlib.pyplot as plt
```

Reading File
    
We will use pandas to read our CSV

```python
FILENAME = 'datasets/articles_data_homeproject.csv'

data = pd.read_csv(FILENAME, index_col=0)
data.head(8)
```

# Check Types
Check types, describe columns and review missing values

```python
data.isnull().sum().to_frame()
```

```python
COLUMNS = [
        'source_name',
        'author',
        'engagement_reaction_count',
        'engagement_comment_count',
        'engagement_share_count',
        'engagement_comment_plugin_count'
        ]
data = data[COLUMNS].dropna(axis=0, how='any')
```

```python
egagement_reaction_count = data.groupby('source_name')['engagement_reaction_count'].sum()
egagement_reaction_count.to_frame()
```

```python
highest_source = data['source_name'].value_counts()
highest_source.to_frame()
```

# Data Visualization

```python
# Create list based on above frame
labels = list(highest_source.index.values)

plt.pie(highest_source, labels=labels, autopct='%1.1f%%')
plt.title('Highest amount of news')
plt.show()
```
